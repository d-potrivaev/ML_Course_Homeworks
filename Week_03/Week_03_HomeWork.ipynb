## Plotting with MatPlotLibs

### Ver 1

```python
x = df["SibSp"]
y = df["Age"]
```

```python
plt.plot(x, y, "r.")
plt.xlabel("Number of siblings")
plt.ylabel("Age")
plt.title("Titanic Data")
plt.show()
```

### Ver 2

```python
w = df["PassengerId"]
z = df["Age"]
```

```python
plt.subplot(1, 2, 1) # 1 - кол-во рядов, 2 - кол-во столбиков, 1 - позиция
plt.plot(x, y, "b*")
plt.subplot(1, 2, 2) # 1 - кол-во рядов, 2 - кол-во столбиков, 2 - позиция
plt.plot(w, z, "g>")
```

### Ver 3

```python
fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.45, 0.55, 0.4, 0.3])

ax1.plot(x, y, "b>")
ax1.set_xlabel("Num of siblings")
ax1.set_ylabel("Age")
ax1.set_title("Outside plot")

ax2.plot(w, z, "r--")
ax2.set_xlabel("Pass Id")
ax2.set_ylabel("Age")
ax2.set_title("Inside plot")
```

## Plotting with Seaborn

### Ver 1

```python
sns.set_theme(style="ticks")
sns.lmplot(data=df, x="PassengerId", y="Age", palette="muted", hue="Pclass")
```

### Ver 2

```python
sns.set_theme(style="darkgrid")
_, ax = plt.subplots(figsize=(10,5))
sns.lineplot(data=df, x="SibSp", y="Age", palette="muted", ax=ax, hue="Survived")
```

### Ver 3

```python
sns.displot(data=df, x="Pclass", y="Age")
```

### Ver 4

```python
sns.heatmap(df.pivot_table(values="Age", index="SibSp", columns="Pclass"))
```

### Ver 5

```python
sns.boxplot(data=df, x="Pclass", y="Age", hue="Sex", palette=["m", "g"])
```
